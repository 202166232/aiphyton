```python
elements = [
    '태조', '정종', '태종', '세종', '문종', '단종', '세조', '예종', '성종', '연산군',
    '중종', '인종', '명종', '선조', '광해군', '인조', '효종', '현종', '숙종', '경종',
    '영조', '정조', '순조', '헌종', '철종', '고종', '순종'
]

print(f"조선 왕 리스트: {elements}")
print(f"리스트 총 {len(elements)}명 ")
```

```python
king_number = int(input("몇 대 조선 왕의 이름을 알고 싶으신가요? "))

index = king_number - 1

if 0 <= index < len(elements):
    king_name = elements[index]
    print(f"{king_number}대 조선 왕은 '{king_name}'입니다")
else:
    print(f"{king_number}대 왕은 리스트에 없거나, 유효하지 않은 숫자입니다.")
```
