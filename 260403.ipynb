```python
answer = random.randint(1, 100)
attempts = 0

print("1부터 100 사이의 숫자를 맞춰보세요!")

while True:
    try:
        user_guess = int(input("숫자를 입력하세요: "))
        attempts += 1

        if user_guess < answer:
            print("더 큽니다! 다시 시도하세요.")
        elif user_guess > answer:
            print("더 작습니다! 다시 시도하세요.")
        else:
            print(f" {attempts}번 만에 숫자를 맞췄습니다. 정답은 {answer}였습니다.")
            break
    except ValueError:
        print("유효한 숫자를 입력해주세요.")
```

```python
try:
    num = int(input("약수를 찾을 숫자를 입력하세요: "))
    if num <= 0:
        print("양의 정수를 입력해주세요.")
    else:
        print(f"{num}의 약수:")
        divisors = []
        for i in range(1, num + 1):
            if num % i == 0:
                divisors.append(i)
        print(divisors)
except ValueError:
    print("유효한 정수를 입력해주세요.")
```
