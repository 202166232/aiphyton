```python
import pandas as pd
from sklearn.datasets import load_iris

# Iris 데이터셋 로드
iris = load_iris()

# DataFrame으로 변환
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# 타겟 변수(붓꽃 종류) 추가
# iris.target_names는 ['setosa', 'versicolor', 'virginica'] 입니다.
iris_df['species'] = iris.target_names[iris.target]

print('Iris 데이터셋 로드 완료:')
display(iris_df.head())
```

```python
# 'Iris-versicolor' 종류만 골라서 평균 꽃잎 길이를 구해보세요
versicolor_df = iris_df[iris_df['species'] == 'versicolor']
versicolor_petal_length_mean = versicolor_df['petal length (cm)'].mean()
print(f"'Iris-versicolor'의 평균 꽃잎 길이: {versicolor_petal_length_mean:.2f} cm")

# 'Iris-virginica' 종류의 최대 꽃받침 너비를 찾아보세요
virginica_df = iris_df[iris_df['species'] == 'virginica']
virginica_sepal_width_max = virginica_df['sepal width (cm)'].max()
print(f"'Iris-virginica'의 최대 꽃받침 너비: {virginica_sepal_width_max:.2f} cm")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 7))
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='species', data=iris_df, s=100, alpha=0.8)
plt.title('붓꽃 종류별 꽃잎 길이와 너비 산점도')
plt.xlabel('꽃잎 길이 (cm)')
plt.ylabel('꽃잎 너비 (cm)')
plt.legend(title='종류')
plt.grid(True)
plt.show()

print("\n관찰 결과:")
print("Iris-setosa는 다른 두 종류와 명확하게 구별됩니다.")
print("Iris-versicolor와 Iris-virginica는 어느 정도 겹치는 부분이 있지만, 대체로 Iris-virginica가 더 긴 꽃잎 길이와 넓은 꽃잎 너비를 가집니다.")
```

```python
# 붓꽃 종류별 각 특성의 평균 계산
species_mean = iris_df.groupby('species').mean()
print("붓꽃 종류별 평균:")
display(species_mean)

# 어떤 종류가 꽃받침이 가장 긴가요?
longest_sepal_species = species_mean['sepal length (cm)'].idxmax()
print(f"\n꽃받침이 가장 긴 종류: {longest_sepal_species} (평균 {species_mean['sepal length (cm)'].max():.2f} cm)")

# 어떤 종류가 꽃잎이 가장 넓은가요?
widest_petal_species = species_mean['petal width (cm)'].idxmax()
print(f"꽃잎이 가장 넓은 종류: {widest_petal_species} (평균 {species_mean['petal width (cm)'].max():.2f} cm)")
```

### 문제 15: 새로운 값 계산하기

```python
# 꽃받침 면적 (Sepal Area) 계산: SepalLengthCm * SepalWidthCm
iris_df['sepal_area'] = iris_df['sepal length (cm)'] * iris_df['sepal width (cm)']

# 꽃잎 면적 (Petal Area) 계산: PetalLengthCm * PetalWidthCm
iris_df['petal_area'] = iris_df['petal length (cm)'] * iris_df['petal width (cm)']

# 각 종류별 평균 면적 계산
mean_areas = iris_df.groupby('species')[['sepal_area', 'petal_area']].mean()
print("붓꽃 종류별 평균 면적:")
display(mean_areas)

# 어떤 종류가 꽃받침 면적이 가장 큰가요?
largest_sepal_area_species = mean_areas['sepal_area'].idxmax()
print(f"\n꽃받침 면적이 가장 큰 종류: {largest_sepal_area_species} (평균 면적: {mean_areas['sepal_area'].max():.2f} cm²)")

# 어떤 종류가 꽃잎 면적이 가장 큰가요?
largest_petal_area_species = mean_areas['petal_area'].idxmax()
print(f"꽃잎 면적이 가장 큰 종류: {largest_petal_area_species} (평균 면적: {mean_areas['petal_area'].max():.2f} cm²)")
```

```python
# 꽃받침 너비가 3.0 이상이고 꽃잎 너비가 1.5 이상인 데이터 필터링
filtered_df = iris_df[(iris_df['sepal width (cm)'] >= 3.0) & (iris_df['petal width (cm)'] >= 1.5)]

print("꽃받침 너비 3.0 이상, 꽃잎 너비 1.5 이상인 데이터:")
display(filtered_df.head())
print(f"총 {len(filtered_df)}개의 데이터가 해당 조건에 맞습니다.")

# 그 데이터들은 주로 어떤 종류인가요?
filtered_species_counts = filtered_df['species'].value_counts()
print("\n필터링된 데이터의 붓꽃 종류별 개수:")
print(filtered_species_counts)

if not filtered_species_counts.empty:
    most_common_species = filtered_species_counts.index[0]
    print(f"주로 {most_common_species} 종류입니다.")
else:
    print("해당 조건에 맞는 데이터가 없어 주로 어떤 종류인지 알 수 없습니다.")
```

```python
# 꽃잎 면적 (petal_area) 기준으로 내림차순 정렬하여 top 10 데이터 추출
top_10_petal_area = iris_df.sort_values(by='petal_area', ascending=False).head(10)

print("꽃잎 면적 기준 Top 10 데이터:")
display(top_10_petal_area)

# top 10에 가장 많이 포함된 종류는?
top_10_species_counts = top_10_petal_area['species'].value_counts()
print("\nTop 10에 포함된 붓꽃 종류별 개수:")
print(top_10_species_counts)

if not top_10_species_counts.empty:
    most_frequent = top_10_species_counts.idxmax()
    print(f"가장 많이 포함된 종류: {most_frequent} ({top_10_species_counts.max()}개)")
else:
    print("Top 10 데이터가 없어 가장 많이 포함된 종류를 알 수 없습니다.")

# 가장 적게 포함된 종류는?
# 전체 붓꽃 종류와 Top 10에 없는 종류를 비교하여 가장 적게 포함된 종류를 찾습니다.
all_species = iris_df['species'].unique()
species_in_top_10 = set(top_10_species_counts.index)

# Top 10에 없는 종류 찾기 (가장 적게 포함된 종류가 될 가능성 있음)
not_in_top_10 = [s for s in all_species if s not in species_in_top_10]

if not_in_top_10:
    print(f"\n가장 적게 (0개) 포함된 종류: {', '.join(not_in_top_10)}")
elif not top_10_species_counts.empty:
    least_frequent = top_10_species_counts.idxmin()
    print(f"가장 적게 포함된 종류 (Top 10 내): {least_frequent} ({top_10_species_counts.min()}개)")
else:
    print("Top 10 데이터가 없어 가장 적게 포함된 종류를 알 수 없습니다.")
```
