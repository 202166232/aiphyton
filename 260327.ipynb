```python
height1 = float(input("첫 번째 학생의 키를 입력하세요 (cm): "))
height2 = float(input("두 번째 학생의 키를 입력하세요 (cm): "))

print(f"첫 번째 학생의 키: {height1} cm")
print(f"두 번째 학생의 키: {height2} cm")
```

```python
if height1 > height2:
    difference = height1 - height2
    print(f"첫 번째 학생이 두 번째 학생보다 {difference} cm 크다.")
else:
    difference = height2 - height1
    print(f"두 번째 학생이 첫 번째 학생보다 {difference} cm 크다.")
```

```python
score = int(input("학생의 점수를 입력하세요: "))

print(f"입력된 점수: {score}점")


if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
else:
    grade = 'D'

print(f"점수는 {grade}입니다.")
```

```python
y = int(input("연도를 입력하시오: "))

if (y % 400 == 0) or ((y % 4 == 0) and (y % 100 != 0)):
    print("윤년")
else:
    print("평년")
```

```python
import random

user = input("가위, 바위, 보: ")


choices = ['가위', '바위', '보']
computer = random.choice(choices)

print(f"사용자: {user}")
print(f"컴퓨터: {computer}")

!
if user == computer:
    print('결과: 무승부')
elif (user == '가위' and computer == '보') or \
     (user == '바위' and computer == '가위') or \
     (user == '보' and computer == '바위'):
    print('결과: 승리')
else:
    print('결과: 패배')
```
